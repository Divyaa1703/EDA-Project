## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Remove missing values and create a copy to avoid SettingWithCopyWarning
df = df.dropna().copy()

# Clean column names (remove spaces)
df.columns = df.columns.str.strip()

# Convert date columns to datetime, allowing pandas to infer format
df['Scheduled Delivery Date'] = pd.to_datetime(df['Scheduled Delivery Date'], errors='coerce')
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')

# Create Delivery Delay column
df['Delivery_Delay'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# Convert numeric columns
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')

# Create Status column
df['Status'] = df['Delivery_Delay'].apply(lambda x: 'Late' if x > 0 else 'On Time')
```

#### Chart - 3

```python
import numpy as np

plt.figure(figsize=(8,6))

x = df['Weight (Kilograms)']
y = df['Delivery_Delay']

# Combine x and y into a temporary DataFrame to drop NaNs concurrently
temp_df = pd.DataFrame({'x': x, 'y': y}).dropna()
x_cleaned = temp_df['x']
y_cleaned = temp_df['y']

plt.scatter(x_cleaned, y_cleaned, alpha=0.5)

# trend line
# Check if x_cleaned has enough unique values for polyfit
if len(x_cleaned.unique()) > 1:
    z = np.polyfit(x_cleaned, y_cleaned, 1)
    p = np.poly1d(z)
    plt.plot(x_cleaned, p(x_cleaned), color='red', linestyle='--') # Added a trend line color and style
else:
    # If all x values are the same, plot a horizontal line at the mean of y_cleaned
    plt.axhline(y=y_cleaned.mean(), color='red', linestyle='--')

plt.title('Weight vs Delivery Delay (Trend Analysis)', fontsize=14, fontweight='bold')
plt.xlabel('Weight (Kilograms)')
plt.ylabel('Delivery Delay')

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

##### 1. Why did you pick the specific chart?

I used a scatter plot because it helps to understand the relationship between two numerical variables, which are shipment weight and delivery delay. It also allows us to see patterns, trends, or any correlation between them.

##### 2. What is/are the insight(s) found from the chart?

I observed that there is no strong relationship between weight and delivery delay. The points are scattered, which means heavier shipments are not necessarily delayed more than lighter ones.

The trend line also shows a weak relationship, indicating that weight is not a major factor affecting delivery delays.

##### 3. Will the gained insights help creating a positive business impact?


Yes, this insight helps the business understand that delivery delays are likely caused by other factors such as shipment mode, location, or logistics issues rather than weight. This can help the company focus on the right areas to improve delivery performance.

#### Chart - 4

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create weight bins
df['Weight Range'] = pd.cut(df['Weight (Kilograms)'],
                           bins=5)

grouped = df.groupby('Weight Range')['Freight Cost (USD)'].mean()

plt.figure(figsize=(8,5))

plt.bar(grouped.index.astype(str), grouped.values)

plt.title('Average Freight Cost by Weight Range', fontsize=14, fontweight='bold')
plt.xlabel('Weight Range')
plt.ylabel('Average Freight Cost')

plt.xticks(rotation=30)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

I used a bar chart because it helps to compare the average freight cost across different weight ranges. Grouping the data into bins makes it easier to understand how cost changes with weight.

##### 2. What is/are the insight(s) found from the chart?

I observed that as the weight range increases, the average freight cost also increases. This shows a clear positive relationship between shipment weight and cost.

##### 3. Will the gained insights help creating a positive business impact?

Yes, this insight helps in cost planning and budgeting. The business can estimate shipping costs based on weight and make better decisions related to pricing, packaging, and logistics.
