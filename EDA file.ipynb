## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Remove missing values and create a copy to avoid SettingWithCopyWarning
df = df.dropna().copy()

# Clean column names (remove spaces)
df.columns = df.columns.str.strip()

# Convert date columns to datetime, allowing pandas to infer format
df['Scheduled Delivery Date'] = pd.to_datetime(df['Scheduled Delivery Date'], errors='coerce')
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')

# Create Delivery Delay column
df['Delivery_Delay'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# Convert numeric columns
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')

# Create Status column
df['Status'] = df['Delivery_Delay'].apply(lambda x: 'Late' if x > 0 else 'On Time')
```

### What all manipulations have you done and insights you found?


First, I checked the dataset and found some missing values, so I removed them using dropna() to make the data clean.

Then I cleaned the column names because some of them had extra spaces. After that, I converted the date columns into proper date format so that I could calculate the delivery delay.

I created a new column called Delivery_Delay to find the difference between scheduled and actual delivery dates. I also converted columns like freight cost and weight into numeric format for proper analysis.

Finally, I created a new column called Status to classify deliveries as "On Time" or "Late".

From this, I understood that:

Some deliveries are delayed

Shipment mode affects delivery time

Cost varies for different shipments

Overall, after cleaning and processing, the dataset became ready for analysis.

#### Chart - 3

```python
import numpy as np

plt.figure(figsize=(8,6))

x = df['Weight (Kilograms)']
y = df['Delivery_Delay']

# Combine x and y into a temporary DataFrame to drop NaNs concurrently
temp_df = pd.DataFrame({'x': x, 'y': y}).dropna()
x_cleaned = temp_df['x']
y_cleaned = temp_df['y']

plt.scatter(x_cleaned, y_cleaned, alpha=0.5)

# trend line
# Check if x_cleaned has enough unique values for polyfit
if len(x_cleaned.unique()) > 1:
    z = np.polyfit(x_cleaned, y_cleaned, 1)
    p = np.poly1d(z)
    plt.plot(x_cleaned, p(x_cleaned), color='red', linestyle='--') # Added a trend line color and style
else:
    # If all x values are the same, plot a horizontal line at the mean of y_cleaned
    plt.axhline(y=y_cleaned.mean(), color='red', linestyle='--')

plt.title('Weight vs Delivery Delay (Trend Analysis)', fontsize=14, fontweight='bold')
plt.xlabel('Weight (Kilograms)')
plt.ylabel('Delivery Delay')

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

To compare delays across countries.
